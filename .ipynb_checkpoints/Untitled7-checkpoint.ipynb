```python
import pandas as pd
import os

# 1. Automatically find the file in your Downloads
folder_path = '/Users/ritiktoksiya/Downloads/'
possible_files = [f for f in os.listdir(folder_path) if 'Chronic_Kidney_Dsease' in f and f.endswith('.csv')]

if not possible_files:
    print("❌ No kidney disease files found in Downloads. Please check the filename!")
else:
    # Pick the first matching file found
    file_path = os.path.join(folder_path, possible_files[0])
    df = pd.read_csv(file_path)
    print(f"✅ Success! Loaded: {possible_files[0]}")
```

```python
df.head()
```

```python
mappings = {
        'Gender': {0: 'Male', 1: 'Female'},
        'Ethnicity': {0: 'Caucasian', 1: 'African American', 2: 'Asian', 3: 'Other'},
        'EducationLevel': {0: 'None', 1: 'High School', 2: "Bachelor's", 3: 'Higher'},
        'SocioeconomicStatus': {0: 'Low', 1: 'Middle', 2: 'High'} # Assumed based on standard patterns
    }

    # 2. Identify all Binary (0/1) columns you listed
binary_cols = [
        'Smoking', 'FamilyHistoryKidneyDisease', 'FamilyHistoryHypertension', 
        'FamilyHistoryDiabetes', 'PreviousAcuteKidneyInjury', 'UrinaryTractInfections', 
        'ACEInhibitors', 'Diuretics', 'Statins', 'AntidiabeticMedications', 
        'Edema', 'HeavyMetalsExposure', 'Diagnosis'
    ]

    # 3. Apply Mappings
for col, map_dict in mappings.items():
    if col in df.columns:
        df[col] = df[col].map(map_dict)

for col in binary_cols:
    if col in df.columns:
        df[col] = df[col].map({0: 'No', 1: 'Yes'})

print("🚀 All columns mapped to original clinical names!")
print(df[['Ethnicity', 'Gender', 'Smoking', 'Diagnosis']].head())

    
```

```python
# Assuming 'df' is your mapped dataframe
output_path = '/Users/ritiktoksiya/Downloads/RenalTrack_Cleaned_Data.csv'

df.to_csv(output_path, index=False)

print(f"✅ Success! Your edited dataset is saved at: {output_path}")
```
